```python
# Pandas visualizations use Matplotlib and Seaborn as a base
# We can use these libraries by themselves for visualizations

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Use np.linspace(min, max, num_points) to generate 100 points
x = np.linspace(-10,10,100)

# Create a sine wave
y = np.sin(x)

# Matplotlib is incredibly versitle 
# We can generate any plot we want 
plt.plot(x,y,marker='*')
plt.show()
```

```python
df = pd.read_csv("../Data/california_housing_train.csv")

x = df['median_income']
y = df['median_house_value']

# Create a scatter plot with a few more options

plt.scatter(x, y, c='lime', marker='D', edgecolors='black')
plt.show()
```

```python
# Catagorical data

education = ["High School", "Bachelors", "Masters", "PHD"]
values = [32, 80, 19, 5]

# Use 3 plots stacked for visualization
# Mulitple variables can be assigned at the same time, they will have the same value
# plt.subplots(row, col)
fig, axs = plt.subplots(3, 1)

axs[0].bar(education, values)
axs[1].scatter(education, values)
axs[2].plot(education, values)

# We can give it a title
plt.suptitle("Distribution of Education")
plt.show()
```

```python
# Seaborn

# Built ontop of matplotlib
# pandas uses seaborn by default
# a bit easier to use

import seaborn as sns

```

```python
# Default Theme

sns.set_theme()

tips = sns.load_dataset('tips')

sns.relplot(
    data=tips,
    x='total_bill',
    y='tip',
    col='time',
    hue='smoker',
    style='sex',
    size='size'
)
```
