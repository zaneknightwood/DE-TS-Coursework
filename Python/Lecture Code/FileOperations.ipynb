```python
# File Operations

# Python working with Files Open

# Open() can be used with
# 'r' - read = opens for read only access, errors if file doesn't exist
# 'w' - write - opens for writing, will create a file if it doesn't exist
# 'a' - append - any text gets appended, will create a file if it doesn't exist
# 'x' - create - create a file if it doesn't exist, error if the file doesn't exist

# Types of data we can read in
# 't' - text data
# 'b' - bytes/binary

# if we don't specify any options, defaults to 'rt' -- open('text.txt')
```

```python
# For file can use the canonical path or the relative path

file_path = "../Data/Pride_and_Prejudice.txt"

f = open(file = file_path, mode = 'rt', encoding = "utf-8")

# Read the whole thing
whole_book = f.read()

# Can be sliced [starting character : ending character]
print(whole_book[:51])

f.close()
```

```python
f = open(file = file_path, mode = 'rt', encoding = "utf-8")

# We can search for a specific character (exclusive, 0-51 reads characters 0-50)
# Read does not load in entire file, only specified characters
f.seek(0)
print(f.read(51))

f.seek(944)
print(f.read(20))

# Remember to always close your file
f.close()

# Windows - End Bytes \r\n
# Linux/Mac - End Bytes \n
```

```python
# There is an easier way

# This always closes the file at end
with open(file = file_path, mode = 'rt', encoding = "utf-8") as f:
    # book = f.read()  --reads whole file
    # lines = f.readlines()  --puts all lines into list
    # line = f.readline()  --reads one line at a time as a string (can be looped through)
    line = f.readline()

#print(type(book))
print(line)
```

```python
# Write a file

output_file = "../Data/output.txt"
# Remember to always use an escape charater '\n'
line1 = "Writing something to my file\n"
lines2_3 = ["This is the 2nd line\n", "This is the 3rd line\n"]
line4 = "This will be the only line\n"

with open(file = output_file, mode = "wt", encoding = "utf-8") as f:
    f.write(line1)

with open(file = output_file, mode = "at", encoding = "utf-8") as f:
    f.writelines(lines2_3)

with open(file = output_file, mode = "wt", encoding = "utf-8") as f:
    f.write(line4)

output_file2 = "../Data/output2.txt"
with open(file = output_file2, mode = "at", encoding = "utf-8") as f:
    f.writelines(lines2_3)
```
